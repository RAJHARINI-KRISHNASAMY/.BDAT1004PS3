# Question 1
    Occupations
    Step 1. Import the necessary libraries
    Step 2. Import the dataset from this https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user .
    Step 3. Assign it to a variable called users
    Step 4. Discover what is the mean age per occupation
    Step 5. Discover the Male ratio per occupation and sort it from the most to the least
    Step 6. For each occupation, calculate the minimum and maximum ages
    Step 7. For each combination of occupation and sex, calculate the mean age
    Step 8. For each occupation present the percentage of women and men


```python
import pandas as pd
from fractions import Fraction
import decimal
url="https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users=pd.read_csv(url,sep='|', error_bad_lines=False)
users.head(5)
```

### Mean age per occupation

```python
occupation=users.groupby(['occupation'])['age'].agg(['mean','count'])
occupation
```

### Male ratio per occupation and sort it from the most to the least

```python
occupation_men=users.groupby(['occupation'])['gender'].apply(lambda x: (x=='M').sum()).reset_index(name='men_count')
occupation=pd.merge(occupation,occupation_men,left_on='occupation',right_on='occupation')
def get_ratio(count,men):
    r=Fraction(men,count)
    return r
occupation['men_ratio']=occupation.apply(lambda x : get_ratio(x['count'],x['men_count']),axis=1 )
occupation.sort_values(by=['men_ratio'])
```

### For each occupation, calculate the minimum and maximum ages

```python
print("minimum and maximum age of an occupation:")
users.groupby(['occupation'], as_index=False)['age'].agg(['min','max'])
```

### For each combination of occupation and sex, calculate the mean age

```python
users.groupby(['occupation','gender'], as_index=False)['age'].mean()
```

### For each occupation present the percentage of women and men

```python
occupation=occupation.drop(['men_ratio'],axis=1)
```

```python
def women_count(count,men):
    women=count-men
    return women
occupation['women_count']=occupation.apply(lambda x: women_count(x['count'],x['men_count']),axis=1)
```

```python
def percent(total,g):
    p=(g/total)*100
    return p
occupation['women_percent']=occupation.apply(lambda x: percent(x['count'],x['women_count']),axis=1)
occupation['men_percent']=occupation.apply(lambda x: percent(x['count'],x['men_count']),axis=1)
occupation
```

### Year 2061? Do we really have data from this year? Create a function to fix it and apply it

```python
import datetime
data['date'] = pd.to_datetime(data['date'])
def set_date(x):
    if x.year > 2020:
        year = x.year - 100
    else:
        year = x.year
    return datetime.date(year,x.month,x.day)
data['date'] = data['date'].apply(set_date)
data.head()
```

# Question 6
    Create a line plot showing the number of marriages and divorces per capita in the
    U.S. between 1867 and 2014. Label both lines and show the legend.
    Don't forget to label your axes!

```python
import matplotlib.pyplot as plt
us = pd.read_csv('us-marriages-divorces-1867-2014.csv', error_bad_lines=False)
mrg=us['Marriages_per_1000'].tolist()
div=us['Divorces_per_1000'].tolist()
year=us['Year'].tolist()
plt.plot(year, mrg,label = "Marriages")
plt.plot(year,div,label = "Divorces")
plt.xlabel('Year')
plt.ylabel('marriages and divorces per capita')
plt.title(' marriages and divorces per capita in the U.S. between 1867 and 2014.')
plt.legend()
plt.show()
```

# Question 7
    Create a vertical bar chart comparing the number of marriages and divorces per
    capita in the U.S. between 1900, 1950, and 2000.
    Don't forget to label your axes!

```python
temp=us.loc[(us['Year'] == 1900)|(us['Year'] == 1950)|(us['Year'] == 2000)]
y=temp['Year'].tolist()
m=temp['Marriages_per_1000'].tolist()
d=temp['Divorces_per_1000'].tolist()
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))
br1 = np.arange(len(m))
br2 = [x + barWidth for x in br1]
plt.bar(br1, m, color ='g', width = barWidth,
        edgecolor ='grey', label ='Marriages')
plt.bar(br2, d, color ='r', width = barWidth,
        edgecolor ='grey', label ='Divorces')
plt.xlabel('Year', fontweight ='bold', fontsize = 15)
plt.ylabel('number of marriages and divorces per capita', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(m))],y)
plt.legend()
plt.show()
```
